## Task 

#### Find the number of 6s hit in Wankhede stadium
#### Find Highest score made to which venue


#### The data has lot of columns. Instead of carring all the columns lets select the relevant columns to solve the problem

```python
ball_subset = ball.select(['MatcH_id','Runs_Scored'])
```

```python
ball_subset.count()
```

```python
ball_subset.show()
```

```python
match_subset = match.select(['match_id','Venue_Name'])
```

```python
match_subset.count()
```

```python
match_subset.show()
```

```python
ball_subset.createTempView("ball_table")
```

```python
match_subset.createTempView("match_table")
```

```python
df_joined = spark.sql("select match_table.Venue_Name, ball_table.Runs_Scored from ball_table left join match_table on ball_table.MatcH_id == match_table.match_id ")
```

```python
df_joined.collect()
```

```python
df_joined.createTempView("score_venue_data")
```

```python
spark.sql("select * from score_venue_data where Venue_Name=='Wankhede Stadium' and Runs_Scored==6").count()
```

```python
spark.sql("select Venue_Name, sum(Runs_Scored) as total_score from score_venue_data group by Venue_Name").show()
```

```python
spark.sql("select Venue_Name, sum(Runs_Scored) as total_score from score_venue_data group by Venue_Name").collect()
```

```python
df_venue_score = spark.sql("select Venue_Name, sum(Runs_Scored) as total_score from score_venue_data group by Venue_Name")
```

```python
df_venue_score.show()
```

```python
df_venue_score.createTempView("venue_score_table")
```

```python
spark.sql("select * from venue_score_table order by total_score desc").show()
```
